```r
library(MPV) #load the textbook data sets library
Delivery<-p8.3
x1<-Delivery$x1
x2<-Delivery$x2
y<-Delivery$y
```

```r
pairs(cbind(y,x1,x2),pch=19)
```

```r
fit.Del12 <- lm(y ~ x1+x2,data=Delivery)
summary(fit.Del12)
```

```r
fit.Del21 <- lm(y ~ x2+x1,data=Delivery)
summary(fit.Del21)
```

```r
library(scatterplot3d)
s3d <- scatterplot3d(x1, x2, y, pch=16, grid=FALSE, main="Plane of best fit", angle=20)
s3d$plane3d(fit.Del12, col='black')
```
